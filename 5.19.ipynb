```python
from pathlib import Path
import pandas as pd
from scipy.stats import chi2_contingency

# 경로 설정
DATA_DIR = Path('/content/type3')
file_path = DATA_DIR / 'channel_purchase.csv'

# 1. 데이터 불러오기
# 실습 환경에서 파일이 없을 때를 대비해 예외 처리를 추가해두면 안전합니다.
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"Error: {file_path} 경로에 파일이 존재하지 않습니다. 경로를 확인해주세요.")
    # 테스트용 임시 데이터 생성 (실제 파일이 없을 때 주석을 해제해 확인해보세요)
    # df = pd.DataFrame({
    #     'channel': ['A']*50 + ['B']*50,
    #     'purchase_yn': ['yes']*30 + ['no']*20 + ['yes']*15 + ['no']*35
    # })

# 2. 데이터 확인
print('[데이터 상위 5행]')
print(df.head(), '\n')

print('[변수 정보]')
df.info()
print('\n')

# 3. 교차표 생성
table = pd.crosstab(df['channel'], df['purchase_yn'])

print('[교차표]')
print(table, '\n')

# 4. 카이제곱 독립성 검정
chi2, p, dof, expected = chi2_contingency(table)

print('[검정 결과]')
print('chi-square statistic:', round(chi2, 4))
print('p-value:', round(p, 6))
print('degrees of freedom:', dof)

# 5. 기대도수 확인
expected_df = pd.DataFrame(expected, index=table.index, columns=table.columns)

print('\n[기대도수]')
print(expected_df)

# 6. 채널별 구매전환율 계산
# purchase_yn이 문자열('yes', 'no') 형태일 경우를 대비해 'yes' 또는 1을 기준으로 계산합니다.
if df['purchase_yn'].dtype == 'object':
    # 'yes' 값을 1로, 나머지를 0으로 변환하여 평균 계산
    conversion_rate = df.groupby('channel')['purchase_yn'].apply(lambda x: (x == 'yes').mean() * 100).sort_values(ascending=False)
else:
    # 이미 1과 0으로 이루어진 숫자형일 경우
    conversion_rate = df.groupby('channel')['purchase_yn'].mean().sort_values(ascending=False) * 100

print('\n[채널별 구매전환율(%)]')
print(conversion_rate.round(2))

# 7. 해석 출력
print('\n[해석]')
if p < 0.05:
    print('p-value가 0.05보다 작으므로, 유입채널과 구매여부는 독립이 아니며 서로 관련이 있다고 해석할 수 있습니다.')
else:
    print('p-value가 0.05 이상이므로, 유입채널과 구매여부의 관련성을 확인하기 어렵습니다.')

# f-string 출력문이 if-else문 밖에 독립적으로 실행되도록 들여쓰기(Indentation) 수정
print(f'가장 높은 구매전환율 채널은 {conversion_rate.index[0]}이며, 전환율은 {conversion_rate.iloc[0]:.2f}%입니다.')
```

```python
from pathlib import Path
import pandas as pd
from scipy.stats import chi2_contingency

# 경로 설정
DATA_DIR = Path('/content/type3/')
file_path = DATA_DIR / 'gender_preference.csv'

# 1. 데이터 불러오기
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"Error: {file_path} 경로에 파일이 존재하지 않습니다. 경로를 확인해주세요.")
    # 테스트용 임시 데이터 구성 (실제 파일이 없을 때 주석을 해제해 확인해보세요)
    # df = pd.DataFrame({
    #     'gender': ['Male']*60 + ['Female']*60,
    #     'content_type': ['Sports']*40 + ['Drama']*20 + ['Sports']*15 + ['Drama']*45
    # })

# 2. 데이터 확인
print('[데이터 상위 5행]')
print(df.head(), '\n')

print('[변수 정보]')
df.info()  # print()로 감싸면 None이 출력되므로 메서드만 단독 호출하는 것이 깔끔합니다.
print('\n')

# 3. 교차표 생성
table = pd.crosstab(df['gender'], df['content_type'])

print('[교차표]')
print(table, '\n')

# 4. 카이제곱 동질성 검정
chi2, p, dof, expected = chi2_contingency(table)

print('[검정 결과]')
print('chi-square statistic:', round(chi2, 4))
print('p-value:', round(p, 6))
print('degrees of freedom:', dof)

# 5. 기대도수 확인
expected_df = pd.DataFrame(expected, index=table.index, columns=table.columns)

print('\n[기대도수]')
print(expected_df)

# 6. 성별별 최다 선호 유형 확인
top_pref = table.idxmax(axis=1)

print('\n[성별별 최다 선호 유형]')
print(top_pref)

# 7. 행 비율(성별 내 비율) 확인
row_ratio = pd.crosstab(df['gender'], df['content_type'], normalize='index') * 100

print('\n[성별 내 콘텐츠 선호 비율(%)]')
print(row_ratio.round(2))

# 8. 해석 출력 (특수 공백 제거 및 들여쓰기 정상화)
print('\n[해석]')
if p < 0.05:
    print('p-value가 0.05보다 작으므로, 성별에 따라 콘텐츠 선호 분포가 다르다고 해석할 수 있습니다.')
else:
    print('p-value가 0.05 이상이므로, 성별에 따른 콘텐츠 선호 분포 차이를 확인하기 어렵습니다.')

# for문이 if-else문 종료 후 독립적으로 실행되도록 정상적인 들여쓰기 적용
for g in top_pref.index:
    print(f'{g}의 최다 선호 콘텐츠는 {top_pref[g]}입니다.')
```
