```python
from scipy import stats
import pandas as pd

① 단일표본 t-검정
H₀: 평균 연봉 = 5000
data = [4800, 5200, 4900, 5100, 5300, 4700, 5000]
t_stat, p_value = stats.ttest_1samp(data, 5000)
print(f"t-statistic: {t_stat}, p-value: {p_value}")

② 독립표본 t-검정
H₀: 두 그룹의 평균이 같다
group_A = [85, 88, 92, 78, 82, 90]
group_B = [75, 78, 80, 72, 76, 74]

등분산성 검정 (Levene)
levene_stat, levene_p = stats.levene(group_A, group_B)
등분산이면 equal_var=True, 아니면 False (Welch's t-test)
t_stat, p_value = stats.ttest_ind(group_A, group_B, equal_var=(levene_p > 0.05))
print(f"t-statistic: {t_stat}, p-value: {p_value}")

③ 대응표본 t-검정
before = [70, 75, 80, 85, 90]
after = [68, 72, 78, 82, 87]
t_stat, p_value = stats.ttest_rel(before, after)
print(f"t-statistic: {t_stat}, p-value: {p_value}")
```
